```python
import os
import threading
import uvicorn
from pyngrok import ngrok
from kaggle_secrets import UserSecretsClient
from huggingface_hub import login

# Retrieve secrets securely from Kaggle Vault
user_secrets = UserSecretsClient()
HF_TOKEN = user_secrets.get_secret("HF_TOKEN")
NGROK_TOKEN = user_secrets.get_secret("NGROK_TOKEN")

# Authenticate services
login(token=HF_TOKEN)
ngrok.set_auth_token(NGROK_TOKEN)

# Terminate any lingering tunnels to avoid collision
ngrok.kill()

# Establish the public tunnel targeting the FastAPI local port
public_tunnel = ngrok.connect(8000)
print("\n=======================================================")
print("🌐 PACE ENGINE IS LIVE!")
print(f"🔗 ACCESS DASHBOARD HERE: {public_tunnel.public_url}")
print("=======================================================\n")
```
